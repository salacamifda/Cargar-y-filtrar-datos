# Cargar y filtrar datos

```python
import pandas as pd
```

```python
filename = '/content/RailsToTrails_National Count Data_week.xlsx'
bike_df = pd.read_excel(filename)
bike_df.head()
```

# Práctica de filtración

## 1. Solo seleccionen los datos después del 31 de julio de 2020.

```python
#bike_df
```

```python
date_filter = bike_df['Week of']. value_counts()
 
date_filter
```

```python
 date_filter = bike_df['Week of'] > '2020-07-27'
 
 date_filter
```

```python
bike_df[date_filter]
```

## 2. Solo seleccionen los datos donde la columna “Change 2019-2020” es superior al 100 % (superior 1).

```python
bike_df['Change 2019-2020']. value_counts()
```

```python
change_filter = bike_df['Change 2019-2020'] > 1

change_filter
```

```python
bike_df[change_filter]
```

## 3. ¡Únanlos! ¿Hubo alguna semana en la segunda mitad del año (después del 31 de julio de 2020) en la que el cambio fuera superior al 100 %?

```python
bike_df = bike_df.loc[ date_filter & change_filter, :]

bike_df
```

No hubo después del 31 de julio de 2020 un cambio que fuera superior al 100%
